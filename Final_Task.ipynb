```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
data=pd.read_csv("C:/Users/ACER/Desktop/ML_Task 01/DS_ML_FinalTask/DS_ML_FinalTask/student-math.csv",sep=";")
data.columns
```

```python
data["Final_Grade"]=data[["G1","G2","G3"]].sum(axis=1)
data.info()
```

```python

le=LabelEncoder()
```

```python
data.iloc[:,[0,1,3,4,5,8,9,10,11,15,16,17,18,19,20,21,22,]]=data.iloc[:,[0,1,3,4,5,8,9,10,11,15,16,17,18,19,20,21,22,]].apply(lambda col:le.fit_transform(col))
data.head()
```

```python
real_y=data.iloc[:,33].values
real_x=data.iloc[:,0:32].values
```

```python
train_x,test_x,train_y,test_y=train_test_split(real_x,real_y,test_size=0.3,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(train_x,train_y)
```

```python
regressor.score(test_x,test_y)
```

```python
prediction=regressor.predict(test_x)
from sklearn.metrics import r2_score
r2_score(test_y,prediction)
```
